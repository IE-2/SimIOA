```python
def convert_mol_ratio_solvent_ratio(mol_ratio, smiles_list, densities=None):
    """
    mol_ratio: list of mol ratio of each component
    densities: list of densities of each component
    """
    if densities is None:
        densities = [1.0 for i in range(len(mol_ratio))]
    else:
        assert len(mol_ratio) == len(densities) == len(smiles_list)
    mw_list = [rdMolDescriptors.CalcExactMolWt(Chem.MolFromSmiles(smiles)) for smiles in smiles_list]
    mass_list = [mw * mol for mw, mol in zip(mw_list, mol_ratio)]
    volume_list = [mass / density for mass, density in zip(mass_list, densities)]
    volume_ratio = [str(vol) + ':' for vol in volume_list]
    return mw_list, volume_ratio
```
